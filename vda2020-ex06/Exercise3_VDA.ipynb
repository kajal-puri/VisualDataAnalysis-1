```
import pandas as pd
df = pd.read_excel("/content/Data_Cortex_Nuclear.xls")
```

```
missing = df.isnull().sum()
```

```
df = df.interpolate(method = 'linear', limit_direction='forward')
df.isnull().sum()
```

```
df['class'].value_counts()
```

```
csc = df.loc[df['class'] == 'c-CS-s']
csc.count()
```

```
tsc = df.loc[df['class'] == 't-CS-s']
tsc.count()
```

```
df_csc = csc[['pPKCG_N', 'pP70S6_N', 'pS6_N', 'pGSK3B_N', 'ARC_N','class']]
df_tsc = tsc[['pPKCG_N', 'pP70S6_N', 'pS6_N', 'pGSK3B_N', 'ARC_N','class']]
df_big = df_csc.append(df_tsc, ignore_index=True)
df_big['class'].replace("c-CS-s",0,inplace=True)
df_big['class'].replace("t-CS-s",1,inplace=True)
df_big
```

```
import plotly.graph_objects as go
fig = go.Figure(data=
    go.Parcoords(
        line = dict(color = df_big['class'],
                   colorscale = [[0,'gold'],[1,'lightseagreen']]),
        dimensions = list([
            dict(range = [0,4],
                constraintrange = [0,3.8],
                label = 'pPKCG_N', values = df_big['pPKCG_N']),
            dict(range = [0,1],
                label = 'pP70S6_N', values = df_big['pP70S6_N']),
            dict(range = [0,0.2],
                label = 'pS6_N', values = df_big['pS6_N']),
            dict(range = [0,0.4],
                label = 'pGSK3B_N', values = df_big['pGSK3B_N']),
            dict(range = [0,0.2],
                label = 'ARC_N', values = df_big['ARC_N'])
        ])
    )
)

fig.update_layout(
    plot_bgcolor = 'white',
    paper_bgcolor = 'white'
)

fig.show()
```
